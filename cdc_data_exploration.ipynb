```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path="data/PLACES__Place_Data__GIS_Friendly_Format___2022_release.csv"
data_df=pd.read_csv(data_path)
data_df.head(20)
```

```python
florida_df=data_df[data_df.StateAbbr=="FL"]
florida_df
```

```python
fl_df=pd.DataFrame(florida_df[["StateAbbr","PlaceName","TotalPopulation","ACCESS2_CrudePrev","BPHIGH_CrudePrev","CHECKUP_CrudePrev","CHOLSCREEN_CrudePrev","DENTAL_CrudePrev","CERVICAL_CrudePrev"]]
                              )
fl_df.dropna()
fl_df
```

```python
fl_df.sort_values(by=["ACCESS2_CrudePrev","TotalPopulation"],
                 ascending=[False,True]).head(10)
```

```python
y=fl_df["DENTAL_CrudePrev"]
x=fl_df["ACCESS2_CrudePrev"]
width=20
height=15

plt.figure(figsize=(15,10))
plt.title("Dental Care")
plt.xlabel("Percent Uninsured")
plt.ylabel("Percent Received Dental Care in Last Year")
plt.bar(x,y, color='slateblue')

plt.show()
plt.savefig('dental_care.png')
```

```python
y=fl_df["CHECKUP_CrudePrev"]
x=fl_df["ACCESS2_CrudePrev"]
width=20
height=15

plt.figure(figsize=(10,5))
plt.title("Checkups")
plt.xlabel("Percent Uninsured")
plt.ylabel("Percent Received Medical Checkup in Last Year")
plt.bar(x,y, color='slateblue')

plt.show()
```

```python
from scipy.stats import linregress

x=fl_df["CHOLSCREEN_CrudePrev"]
y=fl_df["ACCESS2_CrudePrev"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x* slope + intercept
plt.scatter(x,y)
plt.plot(x,regress_values,"r-")
```
